<strong>Derivada Simbólica definida</strong>

```julia
f(x) = sin(x)*x; f(x)
```

```julia
# primeira derivada

dy = diff(f(x),x,1)
```

```julia
# substituindo x por 2

float(SymPy.subs(dy,x,2)) 
```

Ou mais compacto ainda

```julia
# primeira derivada e substituição de x por 2

float(SymPy.subs(diff(f(x),x,1),x,2))
```

<strong>Derivada Parcial</strong>

```julia
x,y = Sym("x,y")
```

```julia
fxy =  x*y + x^2 -2*y
```

```julia
# Primeira Derivada parcial em relação a variável x

DyDx = diff(fxy,x,1);

DyDx
```

```julia
# Testando para x = 0 e y = 1

DyDx(0,1) 
```

```julia
# Primeira derivada parcial em relação a variável y

DyDx = diff(fxy,y,1);
DyDx
```

```julia
# Testando para x = 0 e y = 1

DyDx(0,1)
```

<strong>Integração Multipla</strong>

Integral Multipla Simbólica Indefinida

```julia
using SymPy

x,y = Sym("x,y")
```

Integral Dupla Indefinida

```julia
f = x*y-x+y;
f
```

$ \int \int {f(x,y)} ~dxdy $

```julia
int_dupla = SymPy.integrate(SymPy.integrate(f,x),y)
```

Ou colocando direto a função no comando

```julia
int_dupla = SymPy.integrate(SymPy.integrate(x*y-x+y,y),x)
```

Ou ainda

```julia
SymPy.integrate(f, (x), (y))
```

Integral Tripla Indefinida

```julia
using SymPy;

x,y,z = Sym("x,y,z")
```

```julia
f = x*y*z-x+y-z;
f
```

$ \int \int \int {f(x,y)} ~dxdydz $

```julia
int_tripla_a = SymPy.integrate(SymPy.integrate(SymPy.integrate(x*y-x+y,x),y),z)
```

```julia
SymPy.simplify(int_tripla_a)
```

```julia
int_tripla_b = SymPy.integrate(f, (x), (y), (z))
```

```julia
SymPy.simplify(int_tripla_b)
```

```julia
int_tripla_a == int_tripla_b
```

Integral Multipla Simbólica Definida

```julia
using SymPy;

x,y= Sym("x,y")
```

Integral Multipla Dupla Definida

```julia
f = x*y-x+y;
f
```

$ \int_{y-1}^{y+1} \int_{x-1}^{x+1} {f(x,y)} ~dxdy $

```julia
int_dupla = SymPy.integrate(SymPy.integrate(f,x,x-1,x+1),y,y-1,y+1)
```

Outro exemplo: 

$ \int_{0}^{1} \int_{0}^{1} {f(x,y)} ~dxdy $

```julia
int_dupla = SymPy.integrate(SymPy.integrate(f,x,0,1),y,0,1)
```

Ou ainda mais prático

```julia
SymPy.integrate(f, (x, 0, 1), (y, 0, 1))
```

Integral Multipla Tripla Definida

```julia
using SymPy
x, y, z = Sym("x,y,z")
```

```julia
f = x*y*z-x+y-z;
f
```

$ \int_{0}^{1} \int_{y-1}^{y+1} \int_{x-1}^{x+1} {f(x,y)} ~dxdydz $

```julia
int_tripla = SymPy.integrate(SymPy.integrate(SymPy.integrate(f,x,x-1,x+1),y,y-1,y+1),z,0,1)
```

Ou ainda mais prático

```julia
SymPy.integrate(f, (x, x-1 , x+1 ), (y, y-1 , y+1 ),(z, 0, 1))
```

Outro exemplo

```julia
f = x*y*z-x+y-z;
f
```

$\int \int \int {f(x,y)} ~dxdydz $

```julia
int_tripla = SymPy.integrate(SymPy.integrate(SymPy.integrate(f,x,0,1),y,0,1),z,0,1);
float(int_tripla)
```

Ou ainda mais prático

```julia
integrate(f, (x, 0, 1), (y, 0, 1),(z,0,1))
```
