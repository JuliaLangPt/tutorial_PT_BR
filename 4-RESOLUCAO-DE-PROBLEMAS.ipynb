<strong>Calculo da equação do espaço S(m) = S0+ V*t</strong>

```julia
# Sendo S0 em metros (m),V em metros por segundo(m/s) e t em segundos (s)

S0 = 2
V = 4
t = 10

S0,V,t
```

```julia
S = S0 +V*t

print("O valor do espaço final S é: ",S,"m")
```
