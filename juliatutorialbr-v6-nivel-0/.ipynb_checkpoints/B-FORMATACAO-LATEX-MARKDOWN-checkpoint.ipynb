# FORMATAÇÃO LATEX E MARKDOWN

Jypyter permite formatar textos e acrescentar formulas, tabelas e outros comandos em `Latex` e `Markdown`. `Markdown` é uma linguagem de marcação criada por John Gruber e Aaron Swartz em 2004 que converte texto em XHTML utlizando html2text, que permite uma formatação mais simples para páginas Web. Dessa forma, é possível criar tabelas, títulos, blocos, códigos, listas, símbolos, etc., de maneira clara e precisa. 
<img src="Figuras/markdown.png" align="center" width="200">

`Latex` é um sistema de marcação semelhante ao `Markdown` desenvolvido na década de 80 por Leslie Lamport, destinado à criação de documentos de alta qualidade tipográfica. Através do `Latex` é possível processar textos para a criação de monografias, artigos, teses e outros. 
<img src="Figuras/latex.png" align="center" width="200">

A melhor forma de acrescentar um símbolo ou uma fórmula num notebook é através de um editor de `Latex`(https://www.codecogs.com/latex/eqneditor.php?lang=pt-br). Por meio de um editor é possível formar uma construção visual e após é possível copiar o código e colar no `markdown` entre `$ e $`. Ex:
```md
$\frac{2}{4} or {2 \over 4}$ 
``` 
Resulta em: $\frac{2}{4}$ 

## BÁSICO DO MARKDOWN

+ **Capítulos**

`# primeiro nivel`
# primeiro nível 

`## segundo nivel` 
## segundo nível 

`### terceiro nivel`
### terceiro nivel

$ $

+ **Blocos**

`> A`

`>> B`

`>>> C`


> A
>> B
>>> C

+ **Itálico**

`*palavra itálico*`

*palavra itálico*

`_palavra itálico_`

_palavra itálico_

+ **Negrito**

`**palavra negrito**`

**palavra negrito**

`__palavra negrito__`

__palavra negrito__

* **Nível (use tab para criar o nivel)**

``` 
* nivel 1
    * nivel 2
       * nivel 3
```
