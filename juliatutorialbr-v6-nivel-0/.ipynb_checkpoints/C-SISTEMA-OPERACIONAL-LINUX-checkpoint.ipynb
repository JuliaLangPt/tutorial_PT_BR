# SISTEMA OPERACIONAL LINUX

## O QUE É LINUX?
<img src="Figuras/tux.png" align="center" width="100"> 
<center>Fonte: commons.wikimedia.org</center>

Sistema operacional é um software ou um conjunto de softwares cuja função é gerenciar os recursos do hardware e  fornecer uma interface entre o usuário e o computador. O Linux é um Sistema Operacional, semelhante ao Windows® ou MacOS®, destinado a diversas plataformas de hardware e aplicações, que atualmente possui maior popularidade nos servidores e está 100% presente nos supercomputadores do mundo listados no site top500.org. Silva(2010) informa que:

> O Linux é um sistema operacional criado em 1991 por Linus Torvalds na universidade de Helsinki na Finlândia. É um sistema Operacional de código aberto distribuído gratuitamente pela Internet. Seu código fonte é liberado como Free Software (software livre), sob licença GPL, o aviso de copyright do kernel feito por Linus descreve detalhadamente isto e mesmo ele não pode fechar o sistema para que seja usado apenas comercialmente .

O Linux é o maior e mais abrangente projeto de software de código aberto do mundo na história da computação e de modo geral é formado por:

- **O kernel do Linux **: O núcleo do sistema operacional, é a parte responsável pelo controle do hardware e gerenciamento do acesso aos outros programas e os recursos do sistema.

- **Softwares utilitários**: Softwares importantes relacionados à manutenção do computador e de seus dispositivos. Ex: Sysinfo, Tar, etc.

- **Um ambiente de trabalho** :  interface gráfica. Ex: KDE, Gnome, LXDE, etc

- **Softwares de aplicações**:  softwares que permitem executar uma ou mais tarefas específicas. Ex: Firefox


## PORQUE LINUX?

* **Estabilidade.** Esta é uma qualidade importante na computação científica - travamentos podem comprometer dados de vários dias de processamento. Neste ponto, Silva (2010) informa que "não há a necessidade de se reiniciar o sistema após modificar a configuração de qualquer periférico ou parâmetros de rede. Somente é necessário reiniciar o sistema no caso de uma instalação interna de um novo periférico, falha em algum hardware (queima doprocessador, placa mãe, etc.)".


* **Segurança.** Há duas principais razões que fazem do linux um sistema seguro: não é um sistema popular nos desktops e ser Unix-Like. Em função de sua baixa popularidade nos desktops em relação ao Windows®, o Linux não é alvo preferencial de ataques. Devido ao fato de ser open-source, permite que as vulnerabilidades descobertas sejam corrigidas rapidamente pela comunidade de desenvolvedores e usuários. Além disto, como é baseado no Unix, foi projetado para ser um sistema operacional multiusuário, limitando os privilégios administrativos apenas ao usuários administrador (usuário root). Dessa forma "a proteção é baseada no controle de acesso para leitura, escrita e execução de acesso para o proprietário, o grupo e outros" (TANENBAUM, 2003)


* **Compatibilidade com Hardware.**



* **Liberdade.**  Liberdade é um dos pilares do desenvolvimento do Linux que é realizado por programadores e colaboradores em várias partes do mundo. São centenas de distribuições (veja lista em https://distrowatch.com/) disponíveis para qualquer tipo de aplicação, arquitetura de hardware e gosto pessoal. Além disto, o usuário pode personalizar sua distribuição conforme suas necessidades.


* **Personalização.** O Linux possui código aberto e isto permite fazer ajustes finos na configuração do sistema para atender uma necessidade específica. Dessa forma, é possível criar uma distribuição própria baseada em uma existente ou compilar uma totalmente nova. Existem diversos ambientes de trabalho (KDE, Gnome, LXDE e outros) que permitem configurar o sistema ao seu gosto pessoal.


* **Baixo custo.** Qualquer pessoa pode baixar e instalar em seu computador uma distribuição Linux sem a necessidade de pagar licenças de uso.  Há ainda algumas distribuições Linux que cobram pelo suporte, como é o caso do Ubuntu, mas ainda assim, o custo é menor quando comparado ao preço da licença do Windows.

## QUAL DISTRIBUIÇÃO LINUX É MELHOR?

<img src="Figuras/linux.jpg" align="center" width="500"> 
<center>Fonte: thehacktoday.com</center>

O Linux possui várias versões diferentes para qualquer tipo de usuário de acordo com sua necessidade. Essas versões são chamadas de distribuições ou "distros". "Uma distro do Linux é um conjunto de componentes que são necessários para alcançar um sistema Linux funcional e os processos necessários para instalar esses componentes para alcançar um sistema em execução" (CASTRO,2016). Existem algumas distros destinada para uso em desktops enquanto outras para servidores, diferentes entre si quanto à facilidade de uso, segurança, ambiente gráfico, robustez e etc. O site `www.distrowatch.com` fornece informações sobre mais de 800 distros Linux e ainda mostra quais distros apresentam maior popularidade. Entre as mais populares, podemos destacar:

* **Arch Linux**

O Arch Linux é uma distribuição GNU/Linux de uso geral desenvolvida independentemente voltada para usuários experientes que segue o modelo de lançamento contínuo (rolling-release). Desenvolvido inicalmente pelo canadense Judd Vinet, é focada  em simplicidade, modernidade, pragmatismo, centralidade no usuário e versatilidade.

* **CentOS**

A distribuição CentOS Linux é a abreviação de **C**ommunity **ENT**erprise **O**perating **S**ystem.  É uma distribuição Linux derivada do Red Hat Enterprise Linux (RHEL) com foco no setor coorporativo e mantida pelo CentOS Project. 

* **Debian (www.debian.org/index.pt.html)**

Uma das primeiras distribuições e uma das mais populares e devido ao seu sucesso, foi utilizada para a criação de outras distribuições, com o Ubuntu, Kali Linux, Metamorphose Linux, entre outras. Foi criada em  1993 por Ian Murdock e atualmente mantida por uma comunidade de voluntários. O Debian é recomendado para os usuários que desejam uma forte estabilidade em estações de trabalho ou servidores.

* **Deepin (www.deepin.org)**

Deepin é uma distribuição chinesa atualmente baseada em Debian que visa com foco na facilidade de uso e elegancia. Possui aplicativos nativos como o Deepin Music, Deepin Movie, Deepin Store, e seu próprio Centro de Controle. Sendo fácil de instalar e usar, o deepin pode ser uma boa alternativa do Windows para uso doméstico e em escritórios (DISTROWATCH).

* **Elementary OS (www.elementary.io/pt_BR)**

O elementary OS é uma distribuição de desktop desenvolvida nos EUA baseada em Ubuntu com foco na usabilidade, beleza do ambiente gráfico e baixo consumo de recurso de hardware. Alguns dos seus recursos mais interessantes incluem um ambiente de desktop personalizado chamado Pantheon e muitos aplicativos personalizados (DISTROWATCH).
        
* **Fedora Linux (www.getfedora.org/pt_BR)**

O Fedora é uma distro desenvolvida pelo Projeto Fedora suportado pela comunidade e de propriedade da Red Hat (DISTROWATCH). O Projeto Fedora foca na inovação, integrando novas tecnologias e recursos. Possui  variações personalizadas chamadas de Fedora Spins, desenvolvidas para jogos, segurança, design, computação científica e robótica.

* **Gentoo (www.gentoo.org)**

O nome Gentoo baseado no pinguim do genero Pygoscelis, o pinguim mais rápido do mundo. O Gentoo Linux é uma metadistribuição Linux americana, versátil e rápida, completamente gratuita, voltada para desenvolvedores e profissionais da rede (DISTROWATCH). Ao contrário da maioria das distribuições Linux, normalmente os pacotes são compilados a partir do código fonte, garantindo assim maior eficiência de processamento.

* **Linux Mint (www.linuxmint.com)**

Linux Mint é uma distribuição irlandesa com desing elegante baseada em Ubuntu, com foco na simplicidade de instalação e utilização. Atualmente é uma das cinco distribuição mais populares no www.distrowatch.com.

* **Manjaro Linux (www.manjaro.org)**

O Manjaro Linux é uma distribuição baseado no Arch Linux desenvolvida na Áustria, França e Alemanha. Apresenta processo de instalação intuitivo, detecção automática de hardware e atualização contínua (rooling release). 


* **NetRunner (www.netrunner.com)**

Distribuição baseada em Debian com uma desktop KDE altamente personalizada com aplicativos extras, codecs multimídia, plugins Flash e Java e uma aparência única. Uma edição "Rolling Release" baseada no Manjaro Linux foi relançada em 2017.

* **OpenSUSE (www.opensuse.org)**

O openSUSE é um projeto desenvolvido na Alemanha de forma colaborativo patrocinado pelo SUSE Linux e outras empresas, que se propõem a ser uma distribuição amigável completa do Linux (DISTROWATHC).  O openSUSE oferece uma ótima experiencia de uso e também uma variedade de ferramentas, como OBS, OpenQA, Kiwi, YaST e OSEM.

* **Red Hat Enterprise Linux (www.redhat.com/pt-br)**

Red Hat Enterprise Linux (RHEL) é uma distribuição Linux desenvolvida pela Red Hat e orientada para o mercado comercial(DISTROWATCH). Além do sistema operacional, a Red Hat fornece todo o suporte e treinamento oficial e o Programa de Certificação Red Hat.

* **Scientific Linux (www.scientificlinux.org)**

O Scientific Linux é uma recompilação do Red Hat Enterprise Linux patrocinada pelo Fermi National Accelerator Laboratory. O projeto visa fornecer um sistema operacional estável, escalável e extensível para computação científica.


* **Slackware (www.slackware.com)**

Slackware é "distribuição de um homem só" com foco em segurança, estabilidade e simplicidade. É mantida desde o início em 1993 por Patrick Volkerding. O Slackware Linux oferece aos usuários experientes um sistema totalmente personalizável para atender desde a estação trabalho até servidores. 

* **Ubuntu (www.ubuntu.com)**

A palavra Ubuntu pode ser traduzido como "humanidade com os outros" ou "sou o que sou pelo que nós somos". A distro é uma distribuição derivada do Debian mantida pela Canonical Ltd e foi lançado em 2004 com a proposta de popularizar o Linux nos desktops. O Kubuntu é uma distribuição derivada do Ubuntu com o ambiente de desktop KDE por padrão.
